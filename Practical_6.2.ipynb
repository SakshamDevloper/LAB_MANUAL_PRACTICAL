```python
def price_change_fuzzy(price):
    if price < 0:
        return "negative"
    else:
        return "positive"


def volume_fuzzy(volume):
    if volume < 50:
        return "low"
    else:
        return "high"

def predict_trend(price_change, volume):

    price = price_change_fuzzy(price_change)
    vol = volume_fuzzy(volume)

    if price == "positive" and vol == "high":
        return "BULLISH"

    elif price == "negative":
        return "BEARISH"

    else:
        return "STABLE"

print(predict_trend(2.5, 80))   
print(predict_trend(-1.2, 40))
print(predict_trend(0.5, 30))  
```
