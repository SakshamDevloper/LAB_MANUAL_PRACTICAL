```python
def trend_fuzzy(x):
    if x > 70:
        return "high"
    elif 30 < x <= 70:
        return "medium"
    else:
        return "low"


def volume_fuzzy(x):
    if x > 1000:
        return "high"
    elif 500 < x <= 1000:
        return "medium"
    else:
        return "low"


def sentiment_fuzzy(x):
    if x > 60:
        return "positive"
    elif 40 < x <= 60:
        return "neutral"
    else:
        return "negative"

def predict_market(trend_value, volume_value, sentiment_value):

    t = trend_fuzzy(trend_value)
    v = volume_fuzzy(volume_value)
    s = sentiment_fuzzy(sentiment_value)

    if t == "high" and v == "high" and s == "positive":
        return "STRONG BULLISH"

    elif t == "low" and s == "negative":
        return "BEARISH"

    elif t == "medium" and s == "neutral":
        return "STABLE"

    else:
        return "MODERATE"

print(predict_market(80, 1200, 75))   
print(predict_market(20, 300, 30))    
print(predict_market(50, 700, 50))    
```
