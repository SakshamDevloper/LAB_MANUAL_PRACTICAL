```python
def dfs():
    jugA, jugB = 4, 2
    start = (0, 0)
    goal = (4, 0)

    stack = [start]         
    visited = set([start])
    parent = {start: None}

    while stack:
        a, b = stack.pop()   

        if (a, b) == goal:
            break

        states = [
            (jugA, b),                 
            (a, jugB),                 
            (0, b),                     
            (a, 0),                     
    
            (min(a + b, jugA), max(0, a + b - jugA)),

            (max(0, a + b - jugB), min(a + b, jugB))
        ]

        for state in states:
            if state not in visited:
                visited.add(state)
                parent[state] = (a, b)
                stack.append(state)

    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = parent[cur]

    path.reverse()

    print("Solution Path (DFS):")
    for step in path:
        print(step)

dfs()
```
