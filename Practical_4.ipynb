```python
prerequisites = {
    'math': [],
    'algebra': ['math'],
    'arithmetic': [],
    'physics': ['algebra'],
    'chemistry': ['math'],
    'biology': ['chemistry'],
    'computer_science': ['algebra']
}

knows = {
    'john': ['arithmetic'],
    'jane': ['algebra'],
    'mary': ['math'],
    'jack': ['chemistry']
}

goals = {
    'john': 'computer_science',
    'jane': 'physics',
    'mary': 'biology',
    'jack': 'computer_science'  
}

def can_learn(student, subject):
    prerequisites_for_subject = prerequisites.get(subject, [])
    return all(prereq in knows.get(student, []) for prereq in prerequisites_for_subject)

def recommend_next_step(student):
    goal = goals.get(student)
    if not goal:
        return None 

    for prerequisite in prerequisites.get(goal, []):
        if prerequisite not in knows.get(student, []):
            return prerequisite

    return goal

print('Recommend next step for John:', recommend_next_step('john'))
print('Recommend next step for Jane:', recommend_next_step('jane'))
print('Recommend next step for Mary:', recommend_next_step('mary'))
print('Recommend next step for Jack:', recommend_next_step('jack'))
```
