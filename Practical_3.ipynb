```python
doctors = [
    {
        "name": "Dr_A",
        "specialization": "General Physician",
        "available": ["10Am", "11Am", "12Pm", "1Pm", "2Pm", "3Pm"],
        "booked": {}
    },
    {
        "name": "Dr_B",
        "specialization": "Pediatrician",
        "available": ["9Am", "10Am", "11Am", "1Pm", "2Pm"],
        "booked": {}
    },
    {
        "name": "Dr_C",
        "specialization": "Dermatologist",
        "available": ["1Pm", "2Pm", "3Pm", "4Pm"],
        "booked": {}
    }
]

while True:
    print("\n-- Doctor Appointment System ---\n")

    print("--- Doctor Appointment Slots ---")
    for doctor in doctors:
        print(f"\nDoctor: {doctor['name']}")
        print("Specialization:", doctor["specialization"])

        print("Available Slots:")
        if doctor["available"]:
            for slot in doctor["available"]:
                print(" ", slot)
        else:
            print("  No slots available")

        print("Booked Slots:")
        if doctor["booked"]:
            for slot, patient in doctor["booked"].items():
                print(f"  {slot} â†’ {patient}")
        else:
            print("  None")

    print("\n1. Book Slot")
    print("2. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        doctor_name = input("Enter Doctor Name (Dr_A / Dr_B / Dr_C): ")

        # Find doctor from list
        selected_doctor = None
        for doc in doctors:
            if doc["name"].lower() == doctor_name.lower():
                selected_doctor = doc
                break

        if not selected_doctor:
            print("Invalid Doctor Name!")
            continue

        patient = input("Enter Patient Name: ")
        slot = input("Enter Slot to Book: ")

        if slot in selected_doctor["available"]:
            selected_doctor["available"].remove(slot)
            selected_doctor["booked"][slot] = patient
            print("Slot booked successfully!")
        else:
            print("Slot not available!")

    elif choice == "2":
        print("Exiting the program.")
        break

    else:
        print("Invalid choice, try again!")
```
