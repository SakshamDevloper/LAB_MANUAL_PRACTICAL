```python
from collections import deque

def bfs():
    jugA, jugB = 4, 2
    start = (0, 0)
    goal = (4, 0)

    queue = deque([start])
    visited = set([start])
    parent = {start: None}

    while queue:
        a, b = queue.popleft()

        if (a, b) == goal:
            break

        states = [
            (jugA, b),                 
            (a, jugB),                 
            (0, b),                     
            (a, 0),                     
           
            (min(a + b, jugA), max(0, a + b - jugA)),

            (max(0, a + b - jugB), min(a + b, jugB))
        ]

        for state in states:
            if state not in visited:
                visited.add(state)
                parent[state] = (a, b)
                queue.append(state)

    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = parent[cur]

    path.reverse()

    print("Solution Path:")
    for step in path:
        print(step)

bfs()
```
