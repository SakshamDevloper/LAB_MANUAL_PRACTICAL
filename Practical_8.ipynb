```python
% Facts 
# Write a valid Prolog program to file (use lowercase predicate names, periods, and proper recursion)
prolog_program = """% Facts
father(a,b).
father(a,c).
father(b,d).
father(b,e).
father(c,f).

% Rules
brother(X,Y) :- 
father(Z,X), 
father(Z,Y), X \\= Y.

cousin(X,Y) :- 
father(Z,X), 
father(W,Y), 
brother(Z,W).

grandson(X,Y) :- 
father(Z,X), 
father(Y,Z).

descendent(X,Y) :- 
father(Y,X).

descendent(X,Y) :- 
father(Z,X), 
descendent(Z,Y).
"""

with open("family.p1", "w") as f:
	f.write(prolog_program)

print("Wrote family.p1")

# Writing family.p1
```
