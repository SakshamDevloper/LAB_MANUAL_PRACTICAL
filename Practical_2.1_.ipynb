```python
def trend(label, x):
    if label == "high":
        return x > 70
    elif label == "medium":
        return 30 < x <= 70
    elif label == "low":
        return x <= 30
    return False


def volume(label, x):
    if label == "high":
        return x > 1000
    elif label == "medium":
        return 500 < x <= 1000
    elif label == "low":
        return x <= 500
    return False


def sentiment(label, x):
    if label == "positive":
        return x > 60
    elif label == "neutral":
        return 40 < x <= 60
    elif label == "negative":
        return x <= 40
    return False

rules = [
    (("trend", "high"), ("volume", "high"), ("sentiment", "positive"), "buy"),
    (("trend", "medium"), ("volume", "medium"), ("sentiment", "neutral"), "hold"),
    (("trend", "low"), ("volume", "low"), ("sentiment", "negative"), "sell"),
    (("trend", "high"), ("volume", "low"), ("sentiment", "neutral"), "hold"),
    (("trend", "low"), ("volume", "high"), ("sentiment", "positive"), "hold"),
]

def predict_trend(trend_value, volume_value, sentiment_value):
    for rule in rules:
        (t_label, t_val), (v_label, v_val), (s_label, s_val), action = rule

        if trend(t_val, trend_value) and volume(v_val, volume_value) and sentiment(s_val, sentiment_value):
            return action

    return "no decision"

def predict_trend(trend_value, volume_value, sentiment_value):
    for rule in rules:
        cond1, cond2, cond3, action = rule

        if (trend(cond1[1], trend_value) and
            volume(cond2[1], volume_value) and
            sentiment(cond3[1], sentiment_value)):
            return action

    return "no_decision"

def print_prediction(trend_value, volume_value, sentiment_value):
    action = predict_trend(trend_value, volume_value, sentiment_value)
    print(
        f"Trend: {trend_value}, Volume: {volume_value}, "
        f"Sentiment: {sentiment_value} -> Action: {action}"
    )

print_prediction(75, 1200, 65)

import heapq

distances = {
    ('a', 'b'): 1,
    ('a', 'c'): 4,
    ('b', 'c'): 2,
    ('b', 'd'): 5,
    ('c', 'd'): 1
}
```
