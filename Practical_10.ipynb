```python
import nltk
import spacy
from textblob import TextBlob
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

nlp = spacy.load("en_core_web_sm")

def lexical_spell_check(text):
    blob = TextBlob(text)
    corrected_text = str(blob.correct())
    return corrected_text

def syntactic_analysis(text):
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    return pos_tags

def semantic_analysis(text):
    doc = nlp(text)
    semantic_info = []

    for token in doc:
        semantic_info.append({
            "word": token.text,
            "lemma": token.lemma_,
            "dependency": token.dep_,
            "pos": token.pos_
        })

    return semantic_info

def spell_check_pipeline(text):

    print("\nOriginal Text:")
    print(text)

    corrected = lexical_spell_check(text)

    print("\nCorrected Text:")
    print(corrected)

    print("\nSyntactic Analysis (POS Tags):")
    syntax = syntactic_analysis(corrected)
    for word, tag in syntax:
        print(word, "->", tag)

    print("\nSemantic Analysis:")
    semantics = semantic_analysis(corrected)

    for item in semantics:
        print(item)

text = input("Enter a sentence: ")

spell_check_pipeline(text)
```
